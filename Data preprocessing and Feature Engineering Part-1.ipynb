### 2.1 Numerical & Categorical with Scikit-Learn

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
train = pd.read_csv('D:\\Assets\\house_prediction_train.csv')
test = pd.read_csv('D:\\Assets\\house_prediction_test.csv')
print("shape of train df = ",train.shape)
print("shape of test df = ",test.shape)
```

```python
X_train=train.drop(columns='SalePrice', axis =1) # deleted one columns
y_train=train['SalePrice']
X_test = test.copy()
print("shape of X_train df = ",X_train.shape)
print("shape of y_train df = ",y_train.shape)
print("shape of X_test df = ",X_test.shape)
```

### Missing value imputation

```python
isnull_sum = X_train.isnull().sum()
isnull_sum
```

```python
# finding the numerical variabe which have missing value
num_vars = X_train.select_dtypes(include = ['int64', 'float64']).columns
num_vars_miss = [var for var in num_vars if isnull_sum[var]>0]
```

```python
num_vars_miss
```

```python
# finding the categorical variabe which have missing value
cat_vars = X_train.select_dtypes(include = ['O']).columns
cat_vars_miss = [var for var in cat_vars if isnull_sum[var]>0]
```

```python
cat_vars_miss
```

```python
# now we have to fill value diffrent different place
# at a time fill all numerical and categorical missing value

num_var_mean = ['LotFrontage']
num_var_median = ['MasVnrArea', 'GarageYrBlt']
cat_vars_mode = ['Alley',
 'MasVnrType',
 'BsmtQual',
 'BsmtCond',
 'BsmtExposure',
 'BsmtFinType1',
 'BsmtFinType2',
 'Electrical',
 'FireplaceQu',]
cat_vars_missing = ['GarageType',
 'GarageFinish',
 'GarageQual',
 'GarageCond',
 'PoolQC',
 'Fence',
 'MiscFeature']
```

```python
num_var_mean_Imputer = Pipeline(steps=[("imputer", SimpleImputer(strategy='mean'))])
num_var_median_Imputer = Pipeline(steps=[("imputer", SimpleImputer(strategy='median'))])
cat_var_mode_Imputer = Pipeline(steps=[("imputer", SimpleImputer(strategy='most_frequent'))])
cat_var_missing_Imputer = Pipeline(steps=[("imputer", SimpleImputer(strategy='constant', fill_value ="Miss_RK"))])
```

```python
preprocessor = ColumnTransformer(transformers=[("mean_imputer", num_var_mean_Imputer, num_var_mean),
                              ("median_imputer", num_var_median_Imputer, num_var_median),
                              ("mode_imputer", cat_var_mode_Imputer, cat_vars_mode),
                              ("missing_imputer",cat_var_missing_Imputer ,cat_vars_missing)])
```

```python
preprocessor.fit(X_train)
```

```python
# what value we get in mean

preprocessor.named_transformers_["mean_imputer"].named_steps["imputer"].statistics_
```

```python
train["LotFrontage"].mean()
```

```python
# what value we get in median

preprocessor.named_transformers_["median_imputer"].named_steps["imputer"].statistics_
```

```python
# what value we get in mode

preprocessor.named_transformers_["mode_imputer"].named_steps["imputer"].statistics_
```

```python
# Now we will transform all mean mode median value into our main data set

X_train_clean = preprocessor.transform(X_train)
X_test_clean = preprocessor.transform(X_test)
X_train_clean
```

```python
preprocessor.transformers_
```

```python
# create data frame all missing value filled
X_train_clean_miss = pd.DataFrame(X_train_clean, columns=num_var_mean+num_var_median+cat_vars_mode+cat_vars_missing)
X_train_clean_miss.head()
```
